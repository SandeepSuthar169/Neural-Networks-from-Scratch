```python
inputs =[1, 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [3.2, 6.8, 6, 0.01]
weights3 = [4.4, 8, -0.5, 0]

bias1 = 2
bias2 = 3
bias3 = 0.5

output=[inputs[0]*weights1[0] +
        inputs[1]*weights1[1] +
        inputs[2]*weights1[2] +
        inputs[3]*weights1[3]+ bias1,

        inputs[0]*weights2[0] +
        inputs[1]*weights2[1] +
        inputs[2]*weights2[2] +
        inputs[3]*weights2[3]+ bias2,

        inputs[0]*weights3[0] +
        inputs[1]*weights3[1] +
        inputs[2]*weights3[2] +
        inputs[3]*weights3[3]+ bias3
        ]


print(output)
```

```python
inputs =[1, 2, 3, 2.5]

w = [[0.2, 0.8, -0.5, 1],
      [3.2, 6.8, 6, 0.01],
      [4.4, 8, -0.5, 0]]

b = [2, 3, 0.5]

output = []

for neuron_w, neuron_b  in zip(w, b):
  neuron_output = 0

  for n_input, w in zip(inputs, neuron_w):
    neuron_output += n_input * w + neuron_b
          # y = m * x + b

  output.append(neuron_output)

print(output)
```

Dense Layer **Class**

```python
import numpy as np
import nnfs
nnfs.init()

inputs = 2
neurons = 4

class Layer_dense:

  def __init__(self, inputs, neurons):
    self.w = 0.01*np.random.randn(inputs, neurons)
    self.b = np.zeros((1, neurons))

  def forward(self, inputs):
    self.output =np.dot(inputs, self.w) + self.b

x, y = spiral_data(samples = 100, classes = 3)

dense1 = Layer_dense(2, 3)
dense1.forward(x)

print(dense1.output[:-1])
```
